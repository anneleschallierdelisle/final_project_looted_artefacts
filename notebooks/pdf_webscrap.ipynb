```python
import os
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, unquote
import re

# --------------------
# Paths
# --------------------
base_url = "https://goam.gov.ye"
pages = [
    "/Looted",
    "/Home/Looted?page=2"
]
output_dir = "../data/raw/pdf_files"
os.makedirs(output_dir, exist_ok=True)

# --------------------
# Fonction pour nettoyer le nom de fichier
# --------------------
def sanitize_filename(name):
    # Décode URL et retire caractères invalides
    name = unquote(name)
    name = re.sub(r'[<>:"/\\|?*]', "_", name)
    name = name.strip()
    return name

# --------------------
# Scrape et téléchargement
# --------------------
pdf_links = []

for page in pages:
    url = urljoin(base_url, page)
    r = requests.get(url)
    r.raise_for_status()
    soup = BeautifulSoup(r.text, "html.parser")
    links = [a["href"] for a in soup.find_all("a", href=True)
             if "/Home/DownloadLostFile/" in a["href"]]
    pdf_links.extend(links)

# Supprimer doublons
pdf_links = list(dict.fromkeys(pdf_links))
print(f"📄 Nombre de PDFs trouvés : {len(pdf_links)}")

# Télécharger les PDFs
for link in pdf_links:
    pdf_url = urljoin(base_url, link)
    pdf_id = link.split("/")[-1]
    response = requests.get(pdf_url)
    response.raise_for_status()

    cd = response.headers.get("Content-Disposition")
    if cd and "filename=" in cd:
        filename = cd.split("filename=")[-1].strip(' "')
    else:
        filename = f"file_{pdf_id}.pdf"

    # Nettoyage du nom
    filename = sanitize_filename(filename)

    file_path = os.path.join(output_dir, filename)
    with open(file_path, "wb") as f:
        f.write(response.content)
    print(f"✅ {filename} téléchargé")
```
