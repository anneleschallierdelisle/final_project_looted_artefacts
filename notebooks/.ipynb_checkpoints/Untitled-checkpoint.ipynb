```python
import os
import fitz  # PyMuPDF
import pandas as pd
from transformers import MarianMTModel, MarianTokenizer
from PIL import Image
import io

# --------------------
# Paths
# --------------------
pdf_path = "../data/raw/0-آثارنا المنهوبة_3.pdf"
pdf_name = os.path.basename(pdf_path)
output_dir_img = "../figures/pdf_images"
csv_output = "../data/processed/pdf_extracted_with_chunks.csv"
os.makedirs(output_dir_img, exist_ok=True)
os.makedirs(os.path.dirname(csv_output), exist_ok=True)

# --------------------
# Translator Ar -> En
# --------------------
model_name = "Helsinki-NLP/opus-mt-ar-en"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

def translate_ar_to_en(text):
    if not text.strip():
        return ""
    tokens = tokenizer(text, return_tensors="pt", truncation=True)
    translated = model.generate(**tokens)
    return tokenizer.decode(translated[0], skip_special_tokens=True)

# --------------------
# Helper : ignore corner images
# --------------------
MAX_PAGE_RATIO = 0.6
def is_in_corner(img, page_width, page_height):
    x0, y0, x1, y1 = img['x0'], img['y0'], img['x1'], img['y1']
    if x0 > 350 and y0 > 750: return True
    if x1 < 250 and y0 > 750: return True
    if x1 < 250 and y1 < 50:  return True
    if x0 > 350 and y1 < 50:  return True
    return False

# --------------------
# Open PDF
# --------------------
doc = fitz.open(pdf_path)
chunks = []
chunks_images = []

PAD_X = 1.5
PAD_Y = 1.5
chunk_counter = 1

for page_number, page in enumerate(doc, start=1):
    words = page.get_text("words")
    words = [(w[0], w[1], w[2], w[3], w[4]) for w in words]
    tables = page.find_tables()

    # -------- Tables -> Chunks --------
    for t_idx, t in enumerate(tables, start=1):
        cells_text = []
        for row in t.cells:
            if len(row) < 2: 
                continue
            x0, y0, x1, y1 = row
            cell_words = [w[4] for w in words
                          if (x0 - PAD_X) <= w[0] <= (x1 + PAD_X)
                          and (y0 - PAD_Y) <= w[1] <= (y1 + PAD_Y)]
            text = " ".join(cell_words).strip()
            cells_text.append({"x0": x0, "x1": x1, "y0": y0, "y1": y1, "text": text})

        # Grouper par ligne
        lines = {}
        for c in cells_text:
            y = round(c["y0"])
            lines.setdefault(y, []).append(c)

        table_dict = {}
        table_dict_en = {}
        for line_cells in lines.values():
            if len(line_cells) < 2:
                continue
            line_cells = sorted(line_cells, key=lambda c: c["x0"])
            value_ar = line_cells[0]["text"]
            key_ar = line_cells[-1]["text"]
            key_en = translate_ar_to_en(key_ar)
            value_en = translate_ar_to_en(value_ar)
            table_dict[key_ar] = value_ar
            table_dict_en[key_en] = value_en

        # Centre du chunk
        xs = [c["x0"] for c in cells_text] + [c["x1"] for c in cells_text]
        ys = [c["y0"] for c in cells_text] + [c["y1"] for c in cells_text]
        centre = (sum(xs)/len(xs), sum(ys)/len(ys)) if xs and ys else (0,0)

        # Ajouter chunk
        chunks.append({
            "chunk_id": chunk_counter,
            "pdf_name": pdf_name,
            "page_number": page_number,
            "table_idx": t_idx,
            "table_dict": table_dict,
            "table_dict_en": table_dict_en,
            "centre": centre,
            "image_paths": []
        })
        chunk_counter += 1

    # -------- Images --------
    for i, img in enumerate(page.get_images(full=True), start=1):
        x0, y0, x1, y1 = img[1:5]
        width, height = x1 - x0, y1 - y0
        if is_in_corner({'x0': x0,'y0': y0,'x1': x1,'y1': y1}, page.rect.width, page.rect.height):
            continue
        if width > page.rect.width * MAX_PAGE_RATIO or height > page.rect.height * MAX_PAGE_RATIO:
            continue

        img_obj = doc.extract_image(img[0])
        im = Image.open(io.BytesIO(img_obj["image"]))
        img_filename = f"{pdf_name.replace('.pdf','')}_page{page_number}_img{i}.png"
        img_path = os.path.join(output_dir_img, img_filename)
        im.save(img_path)

        # Associer image au chunk le plus proche
        if chunks:
            closest_chunk = min(
                [c for c in chunks if c['page_number']==page_number],
                key=lambda c: (((x0+x1)/2 - c['centre'][0])**2 + ((y0+y1)/2 - c['centre'][1])**2)
            )
            closest_chunk['image_paths'].append(img_filename)
        chunks_images.append({
            "pdf_name": pdf_name,
            "page_number": page_number,
            "image_path": img_filename,
            "centre": ((x0+x1)/2, (y0+y1)/2)
        })

# --------------------
# Create DataFrame final
# --------------------
rows_data = []
for c in chunks:
    for k_ar, v_ar in c["table_dict"].items():
        rows_data.append({
            "pdf_name": c["pdf_name"],
            "page_number": c["page_number"],
            "chunk_id": c["chunk_id"],
            "key_ar": k_ar,
            "value_ar": v_ar,
            "key_en": c["table_dict_en"].get(translate_ar_to_en(k_ar), ""),
            "value_en": c["table_dict_en"].get(translate_ar_to_en(v_ar), ""),
            "image_paths": ",".join(c["image_paths"])
        })

df = pd.DataFrame(rows_data)

# --------------------
# Export CSV
# --------------------
df.to_csv(csv_output, index=False, encoding="utf-8-sig")
print(df.head(10))
print(f"\n✅ CSV exporté dans : {csv_output}")
```

```python
import os
import fitz  # PyMuPDF
import pandas as pd
from transformers import MarianMTModel, MarianTokenizer
from PIL import Image
import io

# --------------------
# Paths
# --------------------
pdf_path = "../data/raw/0-آثارنا المنهوبة_3.pdf"
pdf_name = os.path.basename(pdf_path)
output_dir_img = "../figures/pdf_images"
csv_output_tables = "../data/processed/pdf_tables.csv"
csv_output_images = "../data/processed/pdf_images.csv"
os.makedirs(output_dir_img, exist_ok=True)
os.makedirs(os.path.dirname(csv_output_tables), exist_ok=True)

# --------------------
# Translator Ar -> En
# --------------------
model_name = "Helsinki-NLP/opus-mt-ar-en"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

def translate_ar_to_en(text):
    if not text.strip():
        return ""
    tokens = tokenizer(text, return_tensors="pt", truncation=True)
    translated = model.generate(**tokens)
    result = tokenizer.decode(translated[0], skip_special_tokens=True)
    # Nettoyer répétitions
    words = result.split()
    seen = set()
    cleaned = []
    for w in words:
        if w not in seen:
            cleaned.append(w)
            seen.add(w)
    return " ".join(cleaned)

# --------------------
# Helper : ignore corner images
# --------------------
MAX_PAGE_RATIO = 0.6
def is_in_corner(img, page_width, page_height):
    x0, y0, x1, y1 = img['x0'], img['y0'], img['x1'], img['y1']
    if x0 > 350 and y0 > 750: return True
    if x1 < 250 and y0 > 750: return True
    if x1 < 250 and y1 < 50:  return True
    if x0 > 350 and y1 < 50:  return True
    return False

# --------------------
# Open PDF
# --------------------
doc = fitz.open(pdf_path)
chunks = []
chunk_counter = 1

PAD_X = 1.5
PAD_Y = 1.5

for page_number, page in enumerate(doc, start=1):
    words = page.get_text("words")
    words = [(w[0], w[1], w[2], w[3], w[4]) for w in words]
    tables = page.find_tables()

    # -------- Tables -> Chunks --------
    for t_idx, t in enumerate(tables, start=1):
        cells_text = []
        for row in t.cells:
            if len(row) < 2: 
                continue
            x0, y0, x1, y1 = row
            cell_words = [w[4] for w in words
                          if (x0 - PAD_X) <= w[0] <= (x1 + PAD_X)
                          and (y0 - PAD_Y) <= w[1] <= (y1 + PAD_Y)]
            text = " ".join(cell_words).strip()
            cells_text.append({"x0": x0, "x1": x1, "y0": y0, "y1": y1, "text": text})

        # Grouper par ligne
        lines = {}
        for c in cells_text:
            y = round(c["y0"])
            lines.setdefault(y, []).append(c)

        table_dict = {}
        table_dict_en = {}
        for line_cells in lines.values():
            if len(line_cells) < 2:
                continue
            line_cells = sorted(line_cells, key=lambda c: c["x0"])
            value_ar = line_cells[0]["text"]
            key_ar = line_cells[-1]["text"]
            key_en = translate_ar_to_en(key_ar)
            value_en = translate_ar_to_en(value_ar)
            table_dict[key_ar] = value_ar
            table_dict_en[key_ar] = {"key_en": key_en, "value_en": value_en}

        xs = [c["x0"] for c in cells_text] + [c["x1"] for c in cells_text]
        ys = [c["y0"] for c in cells_text] + [c["y1"] for c in cells_text]
        centre = (sum(xs)/len(xs), sum(ys)/len(ys)) if xs and ys else (0,0)

        chunks.append({
            "chunk_id": chunk_counter,
            "pdf_name": pdf_name,
            "page_number": page_number,
            "table_idx": t_idx,
            "table_dict": table_dict,
            "table_dict_en": table_dict_en,
            "centre": centre,
            "rel_img": []
        })
        chunk_counter += 1

    # -------- Images --------
    for i, img in enumerate(page.get_images(full=True), start=1):
        x0, y0, x1, y1 = img[1:5]
        width, height = x1 - x0, y1 - y0
        if is_in_corner({'x0': x0,'y0': y0,'x1': x1,'y1': y1}, page.rect.width, page.rect.height):
            continue
        if width > page.rect.width * MAX_PAGE_RATIO or height > page.rect.height * MAX_PAGE_RATIO:
            continue

        img_obj = doc.extract_image(img[0])
        im = Image.open(io.BytesIO(img_obj["image"]))
        img_filename = f"{pdf_name.replace('.pdf','')}_page{page_number}_img{i}.png"
        img_path = os.path.join(output_dir_img, img_filename)
        im.save(img_path)

        # Associer image au chunk le plus proche
        if chunks:
            chunks_in_page = [c for c in chunks if c['page_number']==page_number]
            if chunks_in_page:
                closest_chunk = min(
                    chunks_in_page,
                    key=lambda c: (((x0+x1)/2 - c['centre'][0])**2 + ((y0+y1)/2 - c['centre'][1])**2)
                )
                closest_chunk['rel_img'].append(img_filename)

# --------------------
# Créer DataFrames
# --------------------
# Tables
rows_data = []
for c in chunks:
    for k_ar, v_ar in c["table_dict"].items():
        en_data = c["table_dict_en"][k_ar]
        rows_data.append({
            "chunk_id": c["chunk_id"],
            "pdf_name": c["pdf_name"],
            "page_number": c["page_number"],
            "table_idx": c["table_idx"],
            "key_ar": k_ar,
            "value_ar": v_ar,
            "key_en": en_data["key_en"],
            "value_en": en_data["value_en"],
            "rel_img": ",".join(c["rel_img"])
        })
df_tables = pd.DataFrame(rows_data)

# Images
rows_images = []
for c in chunks:
    for img_filename in c["rel_img"]:
        rows_images.append({
            "chunk_id": c["chunk_id"],
            "pdf_name": c["pdf_name"],
            "page_number": c["page_number"],
            "img_filename": img_filename
        })
df_images = pd.DataFrame(rows_images)



# --------------------
# INIT
# --------------------
chunks = []
chunks_images = []

# --------------------
# LECTURE PDF
# --------------------
with pdfplumber.open(pdf_path) as pdf:
    reader = PdfReader(pdf_path)
    
    for page_number, (pdfpl_page, pypdf_page) in enumerate(zip(pdf.pages, reader.pages), start=1):
        page_width = pdfpl_page.width
        page_height = pdfpl_page.height

        # =========================
        # 1️⃣ TABLES → CHUNKS (mémoire seulement)
        # =========================
        tables = pdfpl_page.find_tables()
 
        for table_idx, t in enumerate(tables, start=1):
            x0, top, x1, bottom = t.bbox
            centre_x = (x0 + x1)/2
            centre_y = (top + bottom)/2

            chunk = {
                "chunk_id": f"{pdf_name}_page{page_number}_table{table_idx}",
                "page_number": page_number,
                "table_idx": table_idx,
                "centre": (centre_x, centre_y)
            }
            chunks.append(chunk)

        # =========================
        # 2️⃣ IMAGES → EXTRACTION + DF
        # =========================
        images = pdfpl_page.images

        for i, img in enumerate(images, start=1):
            width = img['x1'] - img['x0']
            height = img['y1'] - img['y0']

            if is_in_corner(img, page_width, page_height):
                continue
            if width > page_width * MAX_PAGE_RATIO or height > page_height * MAX_PAGE_RATIO:
                continue

            x_centre_img = (img['x0'] + img['x1']) / 2
            y_centre_img = (img['y0'] + img['y1']) / 2

            # --- extraction PNG ---
            try:
                img_key = list(pypdf_page.images.keys())[i-1]
                img_obj = pypdf_page.images[img_key]
                im = Image.open(io.BytesIO(img_obj.data))
            except (IndexError, OSError):
                continue

            img_filename = f"{pdf_name.replace('.pdf','')}_page{page_number}_img{i}.png"
            img_path = os.path.join(output_dir_img, img_filename)
            im.save(img_path)

            # --- match chunk ---
            chunk_id = None
            table_idx_match = None

            chunks_in_page = [c for c in chunks if c['page_number'] == page_number]
            if chunks_in_page:
                closest_chunk = min(
                    chunks_in_page,
                    key=lambda c: ((x_centre_img - c['centre'][0])**2 + (y_centre_img - c['centre'][1])**2)
                )
                chunk_id = closest_chunk["chunk_id"]
                table_idx_match = closest_chunk["table_idx"]

            # --- DF images ---
            chunks_images.append({
                "img_filename": img_filename,
                "page_number": page_number,
                "chunk_id": chunk_id,
                "table_idx": table_idx_match,
                "centre": (x_centre_img, y_centre_img)
            })

# =========================
# DATAFRAME IMAGES UNIQUEMENT
# =========================
df_images = pd.DataFrame(chunks_images)

# =========================
# EXPORT
# =========================


print(df_images.head())
print("✅ Images PNG + CSV exportés")


# --------------------
# Export CSV
# --------------------
df_tables.to_csv(csv_output_tables, index=False, encoding="utf-8-sig")
df_images.to_csv("../data/processed/pdf_images.csv", index=False, encoding="utf-8-sig")

print(f"✅ Tables exportées : {csv_output_tables}")
print(f"✅ Images exportées : {csv_output_images}")
print(df_tables.head())
print(df_images.head())
```

```python
import os
import re
import fitz  # PyMuPDF
import pandas as pd
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast

# --------------------
# Paths
# --------------------
pdf_path = "../data/raw/0-آثارنا المنهوبة_3.pdf"
pdf_name = os.path.basename(pdf_path)
csv_output_tables = "../data/processed/pdf_tables.csv"
os.makedirs(os.path.dirname(csv_output_tables), exist_ok=True)

# --------------------
# Data Cleaning Arabe
# --------------------
def clean_arabic_text(text):
    """
    Nettoyage simple pour améliorer traduction:
    - réduire répétitions de lettres
    - supprimer mots inutiles
    - enlever espaces multiples
    """
    text = re.sub(r'ة+', 'ة', text)       # simplifie répétitions de ة
    text = re.sub(r'مز', '', text)        # supprime 'مز' si non pertinent
    text = re.sub(r'\s+', ' ', text)      # normalise espaces
    return text.strip()

# --------------------
# Translator Ar -> En (MBart50)
# --------------------
model_name = "facebook/mbart-large-50-many-to-many-mmt"
tokenizer = MBart50TokenizerFast.from_pretrained(model_name)
model = MBartForConditionalGeneration.from_pretrained(model_name)

def translate_ar_to_en(text):
    if not text.strip():
        return ""
    text = clean_arabic_text(text)
    tokenizer.src_lang = "ar_AR"
    encoded = tokenizer(text, return_tensors="pt", truncation=True)
    generated_tokens = model.generate(
        **encoded,
        forced_bos_token_id=tokenizer.lang_code_to_id["en_XX"]
    )
    return tokenizer.decode(generated_tokens[0], skip_special_tokens=True)

# --------------------
# Open PDF et création chunks
# --------------------
doc = fitz.open(pdf_path)
chunks = []
chunk_counter = 1
PAD_X = 1.5
PAD_Y = 1.5

for page_number, page in enumerate(doc, start=1):
    words = page.get_text("words")
    words = [(w[0], w[1], w[2], w[3], w[4]) for w in words]
    tables = page.find_tables()

    for t_idx, t in enumerate(tables, start=1):
        cells_text = []
        for row in t.cells:
            if len(row) < 2: 
                continue
            x0, y0, x1, y1 = row
            cell_words = [w[4] for w in words
                          if (x0 - PAD_X) <= w[0] <= (x1 + PAD_X)
                          and (y0 - PAD_Y) <= w[1] <= (y1 + PAD_Y)]
            text = " ".join(cell_words).strip()
            cells_text.append({"x0": x0, "x1": x1, "y0": y0, "y1": y1, "text": text})

        # Grouper par ligne
        lines = {}
        for c in cells_text:
            y = round(c["y0"])
            lines.setdefault(y, []).append(c)

        table_dict = {}
        table_dict_en = {}
        for line_cells in lines.values():
            if len(line_cells) < 2:
                continue
            line_cells = sorted(line_cells, key=lambda c: c["x0"])
            value_ar = line_cells[0]["text"]
            key_ar = line_cells[-1]["text"]
            key_en = translate_ar_to_en(key_ar)
            value_en = translate_ar_to_en(value_ar)
            table_dict[key_ar] = value_ar
            table_dict_en[key_ar] = {"key_en": key_en, "value_en": value_en}

        chunks.append({
            "chunk_id": chunk_counter,
            "pdf_name": pdf_name,
            "page_number": page_number,
            "table_idx": t_idx,
            "table_dict": table_dict,
            "table_dict_en": table_dict_en
        })
        chunk_counter += 1

# --------------------
# Créer DataFrame
# --------------------
rows_data = []
for c in chunks:
    for k_ar, v_ar in c["table_dict"].items():
        en_data = c["table_dict_en"][k_ar]
        rows_data.append({
            "chunk_id": c["chunk_id"],
            "pdf_name": c["pdf_name"],
            "page_number": c["page_number"],
            "table_idx": c["table_idx"],
            "key_ar": k_ar,
            "value_ar": v_ar,
            "key_en": en_data["key_en"],
            "value_en": en_data["value_en"]
        })

df_tables = pd.DataFrame(rows_data)

# --------------------
# Export CSV
# --------------------
df_tables.to_csv(csv_output_tables, index=False, encoding="utf-8-sig")

print(df_tables.head())
print(f"✅ Tables exportées : {csv_output_tables}")
```
